### <span style='color:Yellow'> 2. Find a list of all movies of the same length.  </span>

```python
stmt = '''
select title, length
from film
where length in (select length from film group by length having count(*) > 1)
order by length desc
'''

print(stmt)
result = db.execute(stmt).fetchall()
result
```

```python
class Film(Base):
    __tablename__ = 'film'
    film_id  = Column(Integer, primary_key=True)
    title = Column(String(100))
    length  = Column(Integer)
    def __str__(self):
        return 'Film id:{0}\n Film title: {1}\n Film length: {2}'.format(self.film_id,self.title,self.length)
```

```python
from sqlalchemy import func
session_stmt = session.query(func.count(Film.title),Film.length).group_by(Film.length).order_by(Film.length)
print(session_stmt)
session_result = session_stmt.all()
session_result
```

```python
class City(Base):
    __tablename__ = 'city'
    city_id = Column(Integer, primary_key=True)
    city = Column(String(50))
    def __str__(self):
        return 'City id:{0}\n City: {1}'.format(self.city_id, self.city)

class Address(Base):
    __tablename__ = 'address'
    address_id  = Column(Integer, primary_key=True)
    city_id =  Column(Integer, ForeignKey('city.city_id'))
    def __str__(self):
        return 'Address id:{0}\n Address city id: {1}'.format(self.address_id, self.city_id)

class Customer(Base):
    __tablename__ = 'customer'
    customer_id  = Column(Integer, primary_key=True)
    first_name = Column(String)
    last_name = Column(String)
    address_id = Column(Integer, ForeignKey('address.address_id'))
    def __str__(self):
        return 'Customer id:{0}\n Customer first name: {1}\n Customer last name: {2}\n Customer address id: {3}'.format(self.customer_id, self.first_name, self.last_name, self.address_id)
```

```python
session_stmt =session.query(Customer,Address).join(Address)
print('\nSession join: ')
print(session_stmt)
```
